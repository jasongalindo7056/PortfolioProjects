```python
#Import csv file on to the dataframe
import pandas as pd 
df = pd.read_csv('storms.csv')
```

```python
#Check for null missing values
df.info()
```

```python
#Read all of the columns in the dataframe
df.columns
```

```python
#Selecting columns for the dataframe, now excluding the null columns
df = df[['name', 'year', 'month', 'day', 'hour', 'lat', 'long', 'status',
       'category', 'wind', 'pressure']]
```

```python
#Percentage of storms per category
df['category'].value_counts(normalize=True)
```

```python
#Total storms per year
df.groupby(['year']).count()['name'].plot()
```

```python
#Average storm category per year
df.groupby(['year']).mean()['category'].plot()
```

```python
#Average wind per year
df.groupby(['year']).mean()['wind'].plot()
```

```python
#Save updated dataframe to a new csv file
df.to_csv('Storms-Data-Analysis.csv')
```
