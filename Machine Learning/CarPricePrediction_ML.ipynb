```python
#Get dummy variables for categorical data
df = pd.get_dummies(df, columns=["fuelsystem","aspiration", "carbody","fueltype","drivewheel"], drop_first=True)
```

```python
#Check column names
df.columns
```

```python
#Target Variable 
Y = df["price"]
```

```python
#Features
X = df[['carlength', 'carwidth', 'carheight',
       'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg',
    'fuelsystem_2bbl', 'fuelsystem_4bbl', 'fuelsystem_idi',
       'fuelsystem_mfi', 'fuelsystem_mpfi', 'fuelsystem_spdi',
       'fuelsystem_spfi', 'aspiration_turbo',
       'drivewheel_fwd', 'drivewheel_rwd']]
```

```python
#Split data for training and testing
X_Train, X_Test, y_train, y_test = train_test_split(X, Y , test_size = 0.4,random_state = 101)
```

```python
#Create figure(empty canvas)
lm = LinearRegression()
```

```python
#Train model
lm.fit(X_Train, y_train)
```

```python
#Test model
predictions_on_testing_data = lm.predict(X_Test)
predictions_on_testing_data
```

```python
#Linear Regression between predictions
plt.scatter(y_test, predictions_on_testing_data)
plt.xlabel('real values')
plt.ylabel('predicted values')
plt.show()
```

```python
#Evaluate the model to see if it was effective
MAE = metrics.mean_absolute_error(y_test, predictions_on_testing_data)
MSE = metrics.mean_squared_error(y_test, predictions_on_testing_data)
RMSE = metrics.mean_squared_error(y_test, predictions_on_testing_data,squared=False)
print('MAE:',MAE)
print('MSE:',MSE)
print('RMSE:',RMSE)
```

```python
metrics.explained_variance_score(y_test,predictions_on_testing_data)
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
cdf
```

```python
#Import DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#Create figure (empty canvas)
clf = DecisionTreeRegressor()
```

```python
#Train model
clf = clf.fit(X_Train,y_train)
```

```python
#Test model for predictions
predictions_on_testing_data = clf.predict(X_Test)
predictions_on_testing_data
```

```python
#Evaluate the model to see if it was effective
MAE = metrics.mean_absolute_error(y_test, predictions_on_testing_data)
MSE = metrics.mean_squared_error(y_test, predictions_on_testing_data)
RMSE = metrics.mean_squared_error(y_test, predictions_on_testing_data,squared=False)
print('MAE:',MAE)
print('MSE:',MSE)
print('RMSE:',RMSE)
```

```python
metrics.explained_variance_score(y_test,predictions_on_testing_data)
```

The decision tree regressor is more effective in predicting the car price due to the lower MAE score.
