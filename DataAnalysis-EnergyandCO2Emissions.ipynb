```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('EnergyData1.csv')
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.columns
```

```python
df = df[['Year', 'Energy Related CO2missions (Gigatonnes)',
       'Oil Production (Million barrels per day)',
       'Natural Gas Production (Billion Cubic Metres)',
       'Coal Production (million tons)',
       'Electricity Generation (Terawatt-hours)',
       'Hydroelectricity consumption in TWh',
       'Nuclear energy consumption in TWh']]
```

```python
df.rename(columns={'Energy Related CO2missions (Gigatonnes)': 'Energy_Related_CO2_Emissions',
       'Oil Production (Million barrels per day)' : 'Oil_Production',
       'Natural Gas Production (Billion Cubic Metres)' : 'Natural_Gas_Production',
       'Coal Production (million tons)' : 'Coal_Production',
       'Electricity Generation (Terawatt-hours)' : 'Electricity_Generation',
       'Hydroelectricity consumption in TWh' : 'Hydroelectricity_Consumption',
       'Nuclear energy consumption in TWh' : 'Nuclear_Energy_Consumption'}, inplace=True)
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Oil_Production)
plt.title('Oil Production')
plt.xlabel('Year')
plt.ylabel('Million Barrels Per Day')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Natural_Gas_Production)
plt.title('Natural Gas Production')
plt.xlabel('Year')
plt.ylabel('Billion Cubic Meters')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Coal_Production)
plt.title('Coal Production')
plt.xlabel('Year')
plt.ylabel('Million tons')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Hydroelectricity_Consumption)
plt.title('Hydroelectricity Consumption')
plt.xlabel('Year')
plt.ylabel('Twh')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Nuclear_Energy_Consumption)
plt.title('Nuclear Energy Consumption')
plt.xlabel('Year')
plt.ylabel('Twh')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Hydroelectricity_Consumption)
plt.plot(df.Year, df.Nuclear_Energy_Consumption)
plt.title('Comparison between Hydroelectricity and Nuclear Energy Consumption')
plt.xlabel('Year')
plt.ylabel('TWh')
plt.show()
```
